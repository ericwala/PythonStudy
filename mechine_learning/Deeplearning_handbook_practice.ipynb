```python
# 及閘，反及閘，或閘
a = float(input("w1:"))
b = float(input("w2:"))
c = float(input("theta:"))
def AND(x1, x2):
    w1, w2, theta = a, b, c
    
    tmp = x1*w1 + x2*w2
    if tmp <= theta:
        return 0
    else:
        return 1
print(AND(0,0))
print(AND(0,1))
print(AND(1,0))
print(AND(1,1))

```

```python
# 偏權質與權重
import numpy as np

w1 = float(input("w1:"))
w2 = float(input("w2:"))
theta = float(input("theta:"))
def AND(x1, x2):
    x = np.array([x1, x2])
    w = np.array([w1, w2])
    
    b = theta
    tmp = np.sum(w*x) + b
    
    if tmp <= 0:
        return 0
    else:
        return 1
print(AND(0,0))
print(AND(0,1))
print(AND(1,0))
print(AND(1,1))
```

```python
# 製作互斥及閘
import numpy as np
def AND(x1, x2):#及閘
    x = np.array([x1, x2])
    w = np.array([0.5,0.5])
    b = -0.7
    tmp = np.sum(w*x) + b
    if tmp <= 0:
        return 0
    else:
        return 1
    
def NAND(x1, x2):#反及閘
    x = np.array([x1,x2])
    w = np.array([-0.5,-0.5])
    b = 0.7
    tmp = np.sum(w*x) + b

    if tmp <= 0:
        return 0
    else:
        return 1

def OR(x1, x2):#或閘
    x = np.array([x1,x2])
    w = np.array([0.5,0.5])
    b = -0.2
    tmp = np.sum(w*x) + b
    if tmp <= 0:
        return 0
    else:
        return 1
def XOR(x1,x2):#互斥及閘
    s1 = NAND(x1,x2)
    s2 = OR(x1,x2)
    y = AND(s1,s2)
    return y
print(XOR(0,0))
print(XOR(0,1))
print(XOR(1,0))
print(XOR(1,1))

```
